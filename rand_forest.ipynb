```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/ml_lab/play_tennis.csv')
```

```python
from sklearn.model_selection import train_test_split   
```

```python
df.head()
```

```python
y=df['play']
```

```python
x = df.drop(['play'], axis =1)
```

```python
x
```

```python
x1 = x.drop('day',axis=1)
```

```python
x1
```

```python
x1=pd.get_dummies(x1,columns=['outlook'])
x1=pd.get_dummies(x1,columns=['temp'])
x1=pd.get_dummies(x1,columns=['humidity'])
x1=pd.get_dummies(x1,columns=['wind'])
```

```python
x1
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16,n_jobs=-1, random_state=42)
rnd_clf.fit(x1,y)
```

```python
y_pred = rnd_clf.predict(x1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```
